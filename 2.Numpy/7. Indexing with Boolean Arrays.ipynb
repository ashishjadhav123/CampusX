# Indexing with Boolean Arrays

```python
import numpy as np
```

```python
arr1 = np.random.randint(low=1, high=100, size=20).reshape(4,5)
arr1
```

```python
arr1[1]
```

```python
arr1[0][2]
```

## Indexing using boolean array

```python
arr1[arr1>50]
```

```python
arr1[(arr1 > 50) & (arr1 % 2 != 0)]
```

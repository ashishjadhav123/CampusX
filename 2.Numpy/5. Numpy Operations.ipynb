# Numpy Operations

```python
import numpy as np
```

```python
arr1 = np.array([1,2,3,4,5,6])
arr2 = np.array([4,5,6,7,8,9])
```

```python
arr1 - arr2
```

## vector multiplication

```python
arr1 * arr2
```

## vector to scalar

```python
arr1 > 4
```

```python
arr3 = np.arange(6).reshape(2,3)
arr4 = np.arange(6, 12).reshape(3,2)
```

```python
arr3.dot(arr4)
```

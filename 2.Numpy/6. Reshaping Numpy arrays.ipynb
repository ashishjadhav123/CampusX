# Reshaping Numpy arrays

```python
import numpy as np
```

```python
arr1 = np.arange(24).reshape(6,4)
arr1
```

```python
arr1.ndim
```

```python
arr1.ravel()
```

```python
arr1.transpose()
```

```python
arr2 = np.arange(26,50).reshape(6,4)
arr2
```

```python
np.hstack((arr1, arr2))
```

```python
np.vstack((arr1, arr2))
```

```python
arr1
```

```python
np.hsplit(arr1, 2)
```

```python
np.vsplit(arr1, 6)
```

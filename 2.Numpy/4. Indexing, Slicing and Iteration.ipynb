# Indexing, Slicing and Iteration

```python
import numpy as np
```

```python
arr1 = np.arange(24)
arr1
```

```python
arr1 = np.arange(24).reshape(6, 4)
arr1
```

```python
arr1[1,1]
```

```python
arr1[5,2]
```

```python
arr1[:]
```

```python
arr1[:,2]
```

```python
arr1[:,2:4]
```

```python
arr1[2:4,1:3]
```

```python
arr1[4:,2:]
```

```python
for i in arr1:
    print(i)
```

```python
for i in np.nditer(arr1):
    print(i)
```

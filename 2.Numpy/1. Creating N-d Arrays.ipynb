# Creating N-d Arrays

```python
import numpy as np
```

```python
arr1 = np.array([1,2,3,4,5])
arr1
```

```python
type(arr1)
```

```python
arr2 = np.array([[1,2,3],[3,2,1]])
arr2
```

```python
arr3 = np.zeros((2,3))
arr3
```

```python
arr4 = np.ones((2,3))
arr4
```

```python
arr5 = np.identity(5)
arr5
```

```python
arr6 = np.arange(10)
arr6
```

```python
arr7 = np.arange(0, 50, 2)
arr7
```

```python
arr8 = np.linspace(10, 20, 10)
arr8
```

```python
arr9 = arr8.copy()
arr9
```

## PyPlot API

```python
# Univariate - Numerical
```

```python
data = {
    "Salary" : [25000, 30000, 37000, 28000, 39000, 48000, 55000, 52000, 35000, 26000, 27000, 31000, 33000, 29000, 40000, 36000, 34000, 45000, 47000, 50000]
}
```

```python
df = pd.DataFrame(data=data)
```

```python
df
```

```python
# Line plot

plt.plot(df['Salary'], color='red', marker='o', linestyle = '--')
plt.grid()
plt.show()
```

```python
# Histogram

plt.hist(df['Salary'], bins=5, color='g')
plt.show()
```

```python
# Box polt

plt.boxplot(df['Salary'])
plt.show()
```

```python
df.shape
```

```python
df.loc[20] = 0
```

```python
# Box polt

plt.boxplot(df['Salary'])
plt.show()
```

```python
df.shape
```

```python
df.drop(index = 20, inplace = True)
```

```python
# Univariate: Categorical
```

```python
df["dept"] = ['HR', 'IT', 'Finance', 'HR', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR'] * 2
```

```python
df.info()
```

```python
df.head()
```

```python
# Pie chart

count = df['dept'].value_counts()
```

```python
count
```

```python
plt.pie(count, labels=count.index, autopct='%1.1f')
plt.show()
```

```python
plt.pie(count, labels=count.index, autopct='%1.1f', explode=[0, 0.1, 0])
plt.show()
```

```python
# Count plot

plt.bar(count.index, count, color=['g', 'black', 'r'])
plt.show()
```

```python
 df["Age"] = [22, 25, 29, 24, 30, 35, 40, 36, 28, 22, 23, 25, 26, 24, 31, 29, 27, 33, 35, 38]
```

```python
df.head()
```

```python
# Scatter Plot: 

plt.scatter(x=df['Age'], y=df['Salary'], color='orange')
plt.show()
```

```python
df[df['Salary'] < 10000]
```

```python
df.loc[df['Salary'] < 10000, 'Salary'] = df['Salary'].mean()
```

```python
df[df['Salary'] < 10000]
```

```python
# Scatter Plot: 

plt.scatter(x=df['Age'], y=df['Salary'], color='orange')
plt.show()
```

```python
sort_age = df.sort_values('Age')
```

```python
# Line plot

plt.plot(sort_age['Age'], df['Salary'], marker='*', color='red')
plt.show()
```

```python
# Bar plot

plt.bar(sort_age['Age'], df['Salary'], color='g')
plt.show()
```

```python
# Bivariate 
```

```python
df.columns
```

```python
df.value_counts('dept')
```

```python
hr_sal = df[df['dept'] == 'HR']['Salary']
it_sal = df[df['dept'] == 'IT']['Salary']
fic_sal = df[df['dept'] == 'Finance']['Salary']
```

```python
plt.boxplot([hr_sal, it_sal, fic_sal], labels=['HR', 'IT', 'Finance'])
plt.show()
```

```python
dep_group = df.groupby(['dept'])['Salary'].sum()
dep_group
```

```python
plt.pie(dep_group, labels=dep_group.index, autopct="%1.2f")
plt.show()
```

```python
# Objet Oriented API
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,7))

# Line plot
axs[0].plot(sort_age['Age'], df['Salary'], marker="*", color="red")
axs[0].grid()
axs[0].set_title("Line Plot")
axs[0].plot()

# Histogram
axs[1].hist(df['Salary'])
axs[1].set_title("Histogram")
axs[1].plot()

# Box Plot
axs[2].boxplot(df['Salary'])
axs[2].set_title("Box Plot")
axs[2].plot()

plt.savefig(r"D:\Ashish Doc\Ashish\AI-engineer\Matplotlib\Output\Multiple_Plot.png")
plt.show()
```

```python
# 3 D polt
```

```python
ax = plt.axes(projection = "3d")

ax.scatter(df["Age"], df["Salary"], df["experience"])
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_zlabel("Experience")
plt.show()
```

```python
ax = plt.axes(projection='3d')

ax.scatter(df["Age"], df["Salary"], df["experience"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(data=df_anime, x='rating', y='members')
```

```python
sns.scatterplot(data=df_anime, x='rating', y='members', hue='type')
```

```python
sns.barplot(data=df_anime, x='type', y='rating')
```

```python
sns.boxplot(data=df_anime, x='type', y='rating')
```

```python
plt.figure(figsize=(10,6))

sns.kdeplot(data=df_anime[df_anime['type'] == 'Movie'], x='rating', label='Movie')
sns.kdeplot(data=df_anime[df_anime['type'] == 'TV'], x='rating', label='TV')
sns.kdeplot(data=df_anime[df_anime['type'] == 'OVA'], x='rating', label='OVA')
sns.kdeplot(data=df_anime[df_anime['type'] == 'Special'], x='rating', label='Special')
sns.kdeplot(data=df_anime[df_anime['type'] == 'Music'], x='rating', label='Music')
sns.kdeplot(data=df_anime[df_anime['type'] == 'ONA'], x='rating', label='ONA')

# Improve legend
plt.legend(title="Anime Type")

# Grid for better readability
plt.grid(alpha=0.3)

plt.show()
```

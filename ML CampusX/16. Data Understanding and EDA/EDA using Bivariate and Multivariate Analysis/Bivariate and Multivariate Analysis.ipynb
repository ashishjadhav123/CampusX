# EDA using Bivariate and Multivariate Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.sample(5)
```

```python
titanic = pd.read_csv(r"D:\Ashish Doc\Ashish\AI-engineer\4. Machine-Learning-DS\ML CampusX\3. Data Understanding\EDA using Univariate Analysis\train.csv")
titanic.sample(5)
```

```python
flights = sns.load_dataset('flights')
flights.sample(5)
```

```python
iris = sns.load_dataset('iris')
iris.sample(5)
```

## 1. Scatter Plot = (Numerical - Numerical) 

### Bivariate

```python
# plt.figure(figsize=(10, 8))

sns.scatterplot(data=tips, x='total_bill', y='tip')
```

### Multivariate

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size')

plt.title("Total Bill vs Tip Analysis")
plt.xlabel("Total Bill Amount")
plt.ylabel("Tip Amount")
plt.show()
```

## 2. Bar Plot (Numerical - Categorical)

### Bivariate

```python
sns.barplot(data=titanic, x='Pclass', y='Age')
```

### Multivariate

```python
sns.barplot(data=titanic, x='Pclass', y='Fare', hue='Sex')
```

```python
sns.barplot(data=titanic, x='Pclass', y='Age', hue='Sex')
```

## 3. Box Plot (Numerical - Categorical)

### Bivariate

```python
sns.boxplot(data=titanic, x='Pclass', y='Age')
```

### Multivariate

```python
sns.boxplot(data=titanic, x='Pclass', y='Age', hue='Sex')
```

```python
sns.boxplot(data=titanic, x='Sex', y='Age', hue='Survived')
```

## 4. Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'], hist=False)
sns.distplot(titanic[titanic['Survived'] == 1]['Age'], hist=False)
```

```python
sns.kdeplot(data=titanic[titanic['Survived'] == 0], x='Age', label='Not Survived')
sns.kdeplot(data=titanic[titanic['Survived'] == 1], x='Age', label='Survived')
```

## 5. HeatMap (Categorical - Categorical)

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.info()
```

```python
titanic.groupby('Pclass')['Survived'].mean() * 100
```

```python
(titanic.groupby('Sex')['Survived'].mean() * 100)
```

## 6. ClusterMap (Categorical - Categorical)

```python
pd.crosstab(titanic['Parch'], titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

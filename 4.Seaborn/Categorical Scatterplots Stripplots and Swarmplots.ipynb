# Categorical Scatterplots | Stripplots and Swarmplots

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.catplot(x='day', y='tip', data=data, kind='strip', jitter=0.1)
plt.show()
```

```python
sns.catplot(x='day', y='tip', data=data, kind='swarm')
plt.show()
```

```python
sns.catplot(x='day', y='tip', data=data, kind='swarm', hue='sex')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=data, hue='sex')
```

# Seaborn | Distplots

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Check the type
print(titanic['age'].dtype)

# If it's 'object', convert to numeric first
titanic['age'] = pd.to_numeric(titanic['age'], errors='coerce')

# Now fill the values
titanic['age'] = titanic['age'].fillna(titanic['age'].mean())
```

```python
print(titanic['age'].isnull().sum())
```

```python
sns.displot(titanic['age'], kde=True)
```

```python
sns.displot(data=titanic, x='age', kind='hist', kde=False)
```

```python
sns.distplot(titanic['age'], kde=False, hist=False, rug=True)
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```

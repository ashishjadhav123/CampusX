## The merge function

```python
df_ipl.head()
```

```python
df_deliveries = pd.read_csv(r"D:\Ashish Doc\Ashish\AI-engineer\Pandas\Input\deliveries.csv")
df_deliveries.head()
```

```python
df_input.head()
```

```python
new_df = df_deliveries.merge(df_input, how='inner', left_on='match_id', right_on='id')
```

```python
new_df.columns
```

```python
new_df.groupby(['season','batsman'])['batsman_runs'].sum()
```

#### Multi index series

```python
type(new_df.groupby(['season','batsman'])['batsman_runs'].sum())
```

#### Convert series to Data frame

```python
new_df.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season', keep='first').sort_values('season')
```

```python
new_df.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season', keep='first').sort_values('season')[['season','batsman']]
```
